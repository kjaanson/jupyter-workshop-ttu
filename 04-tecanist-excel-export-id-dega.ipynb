### TECAN-ist andmete eksport

Tavaliselt on harjutud TECAN-ist eksportima andmeid kujul:


|         |         |         |         |     |
| --------|---------|---------|---------|-----|
| well A1 | well A2 | well A3 | well A4 | ... |
| well B1 | well B2 | well B3 | well B4 | ... |
| well C1 | well C2 | well C3 | well c4 | ... |
|   ...   |   ...   |   ...   |   ...   | ... |

Probleem selliste andmete ekspordiga on see et tulevikus kellelgi andmetest aru saada on keeruline. Selle vastu aitab küll hilisem andmete ümber paigutamine Excelis ning töötluste ja teiste tähistuste lisamine kuid see võtab aega ja tihti käsitsi tehes pole ühtne üle erinevate failide.

TECAN-i tarkvara laseb samas igale mõõdetud plaadile lisada igale well-ile kuni kolm ID-d millega koos on siis võimalik eksportida andmeid. Nendeks ID-deks võivad olla siis mõõdetud well-i kohta käiv info. Juhul kui andmeid welli kohta oleks rohkem on võimalik panna ID-ks lihtsalt proovi nimi ja hiljem lisa andmed lisada (vt allpool andmete ühte kokku panemist erinevatest tabelitest).

## DataFrame

`DataFrame` on Pandase üks põhilisi andmestruktuure. Põhimõttelised võib seda võtta nagu Exceli tabelit.

DataFrame saab luua Pandase meetodiga `.DataFrame` edastades sellele *dictionary*.

Andmed on paigutatud Pandase `DataFrames`-is tulpadesse. Andmed on indekseeridud rea kaupa. Kui indeks on määramata siis sellek on lihtsalt rea number.

````python
test_dict = {
    'samplename': ['Sample1A' ,'Sample1B' ,'Sample1C',
                   'Sample2A' ,'Sample2B' ,'Sample2C',
                   'Sample3A' ,'Sample3B' ,'Sample3C',
                   'Sample4A' ,'Sample4B' ,'Sample4C',],
    'konstrukt' : ['plasmiid1','plasmiid1','plasmiid1',
                   'plasmiid1','plasmiid1','plasmiid1',
                   'plasmiid2','plasmiid2','plasmiid2',
                   'plasmiid2','plasmiid2','plasmiid2'],
    't88tlus' :   ['dmso','dmso','dmso',
                   'aine','aine','aine',
                   'dmso','dmso','dmso',
                   'aine','aine','aine'],
    'luc2p' : [ 450, 543, 243,
                456, 644, 453,
               1234,1342,1532,
               4500,6453,4800]}
````

````python
df_simple = pd.DataFrame(test_dict)
df_simple
````

````python
df_columns = pd.DataFrame(test_dict,
                  columns = ['samplename','konstrukt','t88tlus','luc2p'])
df_columns
````

````python
df_withindex = pd.DataFrame(test_dict,
                  columns = ['samplename','konstrukt','t88tlus','luc2p'],
                  index = test_dict['samplename'])
df_withindex
````

Et lisada tehtud DataFrame-ile index-it saab kasutada meetodit `set_index`

````python
df_simple.set_index('samplename')
````

Et võtta välja osasid ridu või tulpi `DataFrame`-st tuleb kasutada `.loc` meetodit. Selleks tuleb edastada sellele Pythoni *list* või *tuple* nimekirjad vastavate ridade või tulpadega mida tahetakse:
```
df.loc[ [... rea indexite nimekiri...] , [tulpade nimede nimekiri] ]
```

````python
df_withindex.loc[['Sample1A', 'Sample2A'],['konstrukt','luc2p']]
````

Asendades ühe või teise kooloniga `:` saab kätte kõik nende ridadega tulbad või siis vastavalt
kõik vastavate tulpadega read.

````python
df_withindex.loc[['Sample1A', 'Sample2A'],:]
````

````python
df_withindex.loc[:,['konstrukt','luc2p']]
````

Teine viis tulpades ja ridades olevatele andmetele ligi pääseda on kasutata `.iloc` meetodit.
Selle meetodiga saab ligi andmetele kasutades tulpade ja ridade numbrilisi indekseid.

````python
df_withindex.iloc[[0, 1],[0,1]]
````

````python
df_withindex.iloc[[0, 1],:]
````

````python
df_withindex.iloc[:,[0, 1]]
````

Läbi `iloc` ja `loc` meetodi saab ka olemasolevates tulpades andmeid muuta.

````python
df_changed = df_withindex.copy()
````

````python
df_changed.loc[["Sample1A"],["luc2p"]] = 10000000
````

````python
df_changed
````

````python
df_withindex
````

Arvutuste tegemine `DataFrame`-iga käib tulpade kaupa.

````python
df_withindex["uusluc2p"] = df_withindex["luc2p"]*10
df_withindex
````

Selle jaoks saab kasutada ka Numpy-st erinevaid matemaatilisi funktsioone. Kõige rohkem meile võivad huvi pakkuda:
  * `np.log`
  * `np.min`
  * `np.max`
  * `np.exp`
  * `np.pow`
  * `np.mean`
  * `np.sdt`

````python
df_withindex["log-luc2p"] = np.log( df_withindex["luc2p"] )
df_withindex
````

`np.min` ja `np.max` funktsioonidega saab kogu tulba miinimumi kasutada arvutustes

````python
df_withindex["mean-centered"] = df_withindex["luc2p"] - np.mean(df_withindex["luc2p"])
df_withindex
````

````python
df_withindex["mean-centered"] = df_withindex["luc2p"] - np.mean(df_withindex["luc2p"])
df_withindex
````

````python
df_withindex["minmax-norm"] = ( df_withindex["luc2p"] - np.min(df_withindex["luc2p"]) ) / ( np.max(df_withindex["luc2p"]) - np.min(df_withindex["luc2p"]) )
df_withindex
````

Et arvutusi teha automaatseks võib ise kirjutada Pythoni funktsioone ja neid kasutada:

````python
def minmax_scale(x):
    return ( x - np.min(x) ) / ( np.max(x) - np.min(x) )
````

````python
df_withindex["minmax-norm-func"] = minmax_scale(df_withindex["luc2p"])
df_withindex
````

## Arvutused gruppidel


Et teha arvutusi pandases gruppidel tuleb kasutada meetodit `.groupby`. Saadud gruppidel saab viia läbi lihtsamaid arvutusi kasutades pandase gruppidesse sisse ehitatud funktsioone `.mean(), std(), min(), .max, count()`. Need funktsioonid kõik tagastavad jällegi Pandase `DataFrame`-i.

### `.mean(), std(), min(), .max, count()`

````python
df_withindex.groupby(['konstrukt']).mean()
````

Grupeerimist saab läbi viia ka mitmel tasemel. Näiteks grupeerida konstrukti ja töötluse järgi ning seejärel arvutada luc2P signaali keskmised:

````python
df_withindex.groupby(['konstrukt','t88tlus']).mean()
````

Pandase grupeeringutesse sisse kirjutatud arvutuste meetodid teevad vastavaid arvutusi kõigil `DataFrame`-i tulpadel. Et saada kätte ennast huvitavad andmed tuleb seega hiljem lihtsalt `.loc` muutujaga või niisama index-iga võtta need välja:

````python
df_withindex.groupby(['konstrukt','t88tlus']).mean()[ ["luc2p","minmax-norm"] ]
````

Saadud `DataFrame`-i võib paigutada uude muutujasse et sellega parem edaspidi manipuleerida oleks:

````python
df_means = df_withindex.groupby(['konstrukt','t88tlus']).mean()[ ["luc2p","minmax-norm"] ]
df_means
````

Praegu saadud `DataFrame`-il on index mitmetasandiline. See võib olla mõndadel juhtudel vajalik kuid et sellest lahti saada saab kasutada `.reset_index()` meetodit:

````python
df_means.reset_index()
````

### `.agg`

Teine viis teha arvutusi gruppidel on gruppide `.agg` funktsioon.

`.agg` käitub natuke erinevalt vastavalt sellele mis talle ette anda.

Andes listi numpy funtsioonidega arvutab ta iga grupi sees vastava funktsiooni tulemuse:

````python
df_withindex.groupby(['konstrukt', 't88tlus']).agg([np.mean, np.std])
````

Andes dictionary funktsioonidega siis kasutatakse dictionary võtmeid tulpadele määratud funktsioonide jaoks. Tagastatud `DataFrame`-il on ainult tulbad mille arvutusi sooviti:

````python
df_withindex.groupby(['konstrukt', 't88tlus']).agg({'luc2p':np.mean})
````

Nii saab dictionarys sisestada ka mitut funktsiooni:

````python
df_withindex.groupby(['konstrukt', 't88tlus']).agg({'luc2p':[np.mean, np.std]})
````

Ja ka ise kirjutatud funktsioone.

````python
def tobe_funktsioon(xgroup):
    """Funktsioon mis lihtsalt paneb grupi siseselt null väärtused"""
    return 0
````

````python
df_withindex.groupby(['konstrukt', 't88tlus']).agg({'luc2p':[np.mean, np.std, tobe_funktsioon]})
````

## Näide - kontrolli vastu normaliseerimine


````python
df_means = df_withindex.groupby(['konstrukt', 't88tlus']).agg({'luc2p':[np.mean]})
df_means
````

````python
df_means.columns = ["luc2p"]
````

````python
df_means
````

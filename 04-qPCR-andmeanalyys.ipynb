```python
dfraw = pd.read_table("example_data_qPCR/test.txt",
                      dtype={"CrossingPoint":np.float64},
                      na_values={"CrossingPoint":[" "]})
```

```python
dfraw = dfraw[["Experiment Name", "Included", "SampleName", "CrossingPoint"]]
```

```python
dfraw["Replicate"] = dfraw["SampleName"].apply(lambda x:x[0])
```

```python
dfraw = dfraw[ dfraw["SampleName"]!="mQ" ]
```

```python
dfraw.dtypes
```

```python
df_means = dfraw.groupby(["Experiment Name", "Replicate", "SampleName"]).agg({"CrossingPoint":[np.mean]}).reset_index()
```

```python
df_means.columns = df_means.columns.droplevel(1)  #Eemaldab grupeerimisega tekkinud nö kahekordsed tulpade nimed
```

```python
df_means.head()
```

```python
df_means["Amplicon"] = df_means["Experiment Name"].apply(lambda x:x.split("-")[1])
```

Võtame välja hprt normaliseeria iga replikaadi sees

```python
df_means = df_means.drop("Experiment Name", axis=1, errors="ignore")
```

```python
df_means.head()
```

```python
df_luc2p = df_means[ df_means.Amplicon == "luc2P" ].set_index(["Replicate", "SampleName"])
```

```python
series_luc2p = df_luc2p["CrossingPoint"]
```

```python
series_luc2p.name
```

```python
df_allmeans = df_means.set_index(["Replicate","SampleName","Amplicon"]).unstack()
```

```python
df_allmeans.columns = df_allmeans.columns.droplevel(0)
```

```python
df_allmeans.columns.name = None
```

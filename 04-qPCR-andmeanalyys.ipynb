```python
dfraw = pd.read_table("example_data_qPCR/test.txt",
                      dtype={"CrossingPoint":np.float64},
                      na_values={"CrossingPoint":[" "]})
```

```python
dfraw = dfraw[["Experiment Name", "Included", "SampleName", "CrossingPoint"]]
```

```python
dfraw["Replicate"] = dfraw["SampleName"].apply(lambda x:x[0])
```

```python
dfraw = dfraw[ dfraw["SampleName"]!="mQ" ]
```

```python
dfraw.dtypes
```

```python
df_means = dfraw.groupby(["Experiment Name", "Replicate", "SampleName"]).agg({"CrossingPoint":[np.mean]}).reset_index()
```

```python
df_means.columns = df_means.columns.droplevel(1)  #Eemaldab grupeerimisega tekkinud nÃ¶ kahekordsed tulpade nimed
```

```python
df_means.head()
```

```python
df_means["Amplicon"] = df_means["Experiment Name"].apply(lambda x:x.split("-")[1])
```
